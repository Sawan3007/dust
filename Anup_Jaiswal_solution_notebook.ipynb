Before starting, I have made some modification to given zip files. putting all images to a single level directory (instaed of /train_0/train/0 , i made it to /train_0/). As train_0 name itself suggest representing a class with 0 lebel. Same goes to train_1 and test folder as well. 

We can use CNN for image classicifcation
Loading required libraries :

```python
import os 
import numpy as np
import pandas as pd 
import cv2
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten,Dense
```

Loading the dataset, according to trian_label.csv , inroder to match the sequence in the final data array

```python
df = pd.read_csv("train_labels.csv")
df.head()
```

```python
image_data   = []
image_label = []

cnt = 0 ;

# laoding images from train_0 folder

for fileName, label in zip(df["file"],df["label"]):
    path = os.path.join("train_0",fileName)
    img = cv2.imread(path)
    if img is not None:
        img = cv2.resize(img,(64,64))
        image_data.append(img)
        image_label.append(label)

    



# loading images from train-1 folder

for fileName, label in zip(df["file"],df["label"]):
    path = os.path.join("train_1",fileName)
    img = cv2.imread(path)
    if img is not None:
        img = cv2.resize(img,(64,64))
        image_data.append(img)
        image_label.append(label)

    



img_array = np.array(image_data)
label_array = np.array(image_label)
```

```python
#Normalizing the images
img_array = img_array/255
```

Spliting the image and label data into 80% (for training the model ) and 20% (fro testing the and evalulating the model)

```python
img_train, img_test, label_train, label_test = train_test_split(img_array,label_array,test_size=0.2,random_state=41)
```

Building CNN Model, why to use CNN ? Because CNN automatically detects spatial data like patterns, edges, textures and shapes. It is mainly used for image classification. 

```python
model = Sequential()

model.add(Conv2D(32,(3,3),activation="relu",input_shape=(64,64,3)))
model.add(MaxPooling2D(2,2))
model.add(Conv2D(64,(3,3),activation="relu"))
model.add(MaxPooling2D(2,2))
model.add(Flatten())
model.add(Dense(128,activation="relu"))
model.add(Dense(1,activation="sigmoid"))
```

```python
model.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
```

Now, Model is complied and ready to use for training. 

```python
model.fit(
    img_train,
    label_train,
    epochs= 10,
    validation_data =(img_test,label_test)
)
```

Predicting on test images and measuring metrices for model evaluation : 

```python
pred_probaility = model.predict(img_test)
pred = (pred_probaility>0.5 ).astype(int) #converting the estimated probability into binary outcomes

#flattening the array
pred = pred.flatten() 
```

Evaluating the Metrics :

```python
accuracy = accuracy_score(label_test, pred)
precision = precision_score(label_test,pred)
recall = recall_score(label_test,pred)
f1_value  = f1_score(label_test,pred)

print("accuracy", accuracy)
print("precision",precision)
print("recall", recall)
print("f1_score",f1_value)
```
